Checks which files containes annotations with some additional texts in brackets.

```python
DIR_PATH = '../../data/6_11_2025_tcc'
```

```python
import os
directories = [d for d in os.listdir(DIR_PATH) if os.path.isdir(os.path.join(DIR_PATH, d))]

directories
```

```python
import os
import textgrid

def has_brackets(string):
    return '(' in string or ')' in string

files_with_brackets = set()

for dir_name in directories:
    dir_path = os.path.join(DIR_PATH, dir_name)
    files = os.listdir(dir_path)


    for file_name in files:
        if not file_name.endswith(".TextGrid"):
            continue

        file_path = os.path.join(dir_path, file_name)

        tg = textgrid.TextGrid.fromFile(file_path)

        tier = tg.getFirst('Asmjeeg')
        for interval in tier:
            text = interval.mark
            if has_brackets(text.strip()):
                files_with_brackets.add(file_path)

print(len(files_with_brackets))
files_with_brackets = sorted(files_with_brackets)
for f in files_with_brackets:
    print('- [ ] ' ,f)


```
